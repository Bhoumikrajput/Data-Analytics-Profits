```python
## Start Time
df.Start_Time
```

```python
df.Start_Time[0]
## yy-mm-dd 
```

```python
## Converting into Date Time format
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
## Seperate hours from TimeStamp
df.Start_Time.dt.hour
```

```python
sns.distplot(df.Start_Time.dt.hour, bins = 24, kde=False,norm_hist = True)
```

```python
## A high % of accidents occurs between 6am to 10am (As people going out for work in between this time)
## Next heighest % of accidents is between 3pm to 6pm
```

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins = 7, kde=False,norm_hist = True)
```

```python
## Ques : Is the distribution of accidents by hour the same on weekend as on weekdays
```

```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6 ]
sns.distplot(sundays_start_time.dt.hour, bins = 24, kde=False,norm_hist = True)
```

```python
mondays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 1 ]
sns.distplot(mondays_start_time.dt.hour, bins = 24, kde=False,norm_hist = True)
```

```python
sns.distplot(mondays_start_time.dt.month, bins = 12, kde=False,norm_hist = True)
```

```python
df_2019 = df[df.Start_Time.dt.year == 2019]
df_2019_Bing = df_2019[df_2019.Source=='Bing']
sns.distplot(df_2019.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```
